# **Statistical distributions**
- Plots of the distribution of at least one variable in a dataset. Some of these methods also compute the distributions.

### **Histogram Plot**

```python
plt.style.use('dark_background')

# plot
fig, ax = plt.subplots()
fig.set_size_inches(16, 8.27)
ax.hist(df['country'].head(10), histtype='bar', weights=df['density (km²)'].head(10), linewidth=0.5, edgecolor="white")

# labeling
ax.set_title('World Population Density')
ax.set_xlabel('Country')
ax.set_ylabel('Density (km²)')
plt.show()
```

### **Box Plot**

```python
plt.style.use('dark_background')


# plot
fig, ax = plt.subplots()
fig.set_size_inches(16, 8.27)
ax.boxplot(df['density (km²)'].head(16), patch_artist=True)
# ax.boxplot(df['area (km²)'].head(16), patch_artist=True)

# labeling
ax.set_title('World Population Density')
ax.set_ylabel('Density (km²)')

# ax.set_xticklabels(['Density'])
plt.show()
```

### **Pie Chart**

```python
plt.style.use('dark_background')

# plot
fig, ax = plt.subplots()
fig.set_size_inches(16, 8.27)
ax.pie(df['density (km²)'].head(10), labels=df['country'].tail(10), autopct='%1.1f%%')

# show
plt.show()
```

```python
# for asian country
plt.style.use('dark_background')

# plot
fig, ax = plt.subplots()
fig.set_size_inches(16, 8.27)

# plot te graph
ax.pie(df[df['continent'] == 'Asia']['density (km²)'].head(10), labels=df[df['continent'] == 'Asia']['country'].head(10), autopct='%1.1f%%')

ax.set_title('World Population Density')

# show
plt.show()
```

### **ErrorBar**
Physics Exp. 09 Error Analysis

```python
x = [40, 20, 10, 5, 2.5]
y = [50.5, 24.4, 11.7, 5.55, 2.9]
# find slop from x and y
m = np.polyfit(x, y, 1)[0]
#  find c
c = np.polyfit(x, y, 1)[1]

# find y_the
# y_the = m * x + c
y_the = np.polyval([m, c], x)

# find error
error = np.std(y) / np.sqrt(len(x))

# plot
fig, ax = plt.subplots(2, 1)
fig.set_size_inches(16, 8.27)
ax[0].plot(x, y, 'o-', label='Experimental Data')
ax[0].plot(x, y_the, 'r-', label='Pridicted Data')
ax[0].legend()
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_ylim(0, 10)
ax[0].set_xlim(2, 10)

ax[1].errorbar(x, y, yerr=error, fmt='o-', label='Experimental Data with error', linewidth=2, capsize=6)

ax[1].legend()
ax[1].set_xlabel('x')
ax[1].set_ylabel('y' + 'Error')
ax[1].set_ylim(0, 60)
ax[1].set_xlim(0, 60)


plt.show()
```
