```python
# import libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Import Data
df = pd.read_csv('https://raw.githubusercontent.com/jsdhami/Python-For-Research/main/Data/hourly_ktm_weather_data.csv')

df = pd.DataFrame({'date':df['date'], 'temp':df['temperature_2m'], 'ppt':df['precipitation'], 'windspeed':df['wind_speed_10m']})
df.head()
print(len(df))
```

```python
df = df.tail(25)
df.head()
# print all
# print(df.to_string())
```

```python
# select data column
x = df['date']
y = df['temp']
z = df['ppt']
w = df['windspeed']
```

```python
# basic rules
fig, ax = plt.subplots()
fig.set_size_inches(16, 6)

ax.plot(x, y, '--', label='Temp.')
ax.plot(x, w, 'o-', label='Wind Speed')

# label
ax.set_ylabel("value")
ax.set_xlabel("date")
ax.set_title("Weather Data of KTM")

# tick labels
# set limit

# grid
plt.grid(True)

# legend
plt.legend()

# show plot
plt.show()

```

```python
# plot
fig, ax = plt.subplots()

fig.set_size_inches(16, 6)

ax.plot(x, y, label='Temperature')

ax.plot(x, z, label='Precipitation')

ax.plot(x, w, label='Wind Speed')

# set tick and tick labels
ax.set_xticks(x[::1])
ax.set_xticklabels(x[::1], rotation=90)

# set labels
ax.set_xlabel('Date')
ax.set_ylabel('Frequency')
ax.set_title('Hourly KTM Weather Data')

plt.legend()

plt.grid(True)

plt.show()
```

```python
# scatter plot

fig, ax = plt.subplots()

fig.set_size_inches(16, 6)

ax.scatter(x, y, label='Temperature')
ax.scatter(x, z, label='Precipitation')
ax.scatter(x, w, label='Wind Speed')

# set tick and tick labels
ax.set_xticks(x[::1])
ax.set_xticklabels(x[::1], rotation=90)

# set labels
ax.set_xlabel('Date')
ax.set_ylabel('Frequency')
ax.set_title('Hourly KTM Weather Data')

plt.legend()

plt.grid(True)

plt.show()
```

```python
# plot
fig, ax = plt.subplots()
fig.set_size_inches(16, 6)

ax.plot(x, y, 'o-', label="Temp.")
ax.plot(x, z, 'o-', label="PPt.")
ax.plot(x, w, 'o-', label = "Wind Speed")

# set tick and tick labels
ax.set_xticks(x[::1])
ax.set_xticklabels(x[::1], rotation=90)

# set labels
ax.set_xlabel('Date')
ax.set_ylabel('Frequency')
ax.set_title('Hourly KTM Weather Data')

plt.grid(True)
plt.legend()
plt.show()

```

```python
# Bar Plot
fig, ax = plt.subplots()
fig.set_size_inches(16, 6)

ax.bar(df['date'], df['temp'], label="Temp.")
# ax.bar(x, w, label = "Wind Speed")
# ax.bar(x, z, label="PPt.")

# set tick and tick labels
ax.set_xticks(x[::1])
ax.set_xticklabels(x[::1], rotation=90)

plt.grid(False)
plt.legend()
plt.show()
```

```python
# stem plot
fig, ax = plt.subplots()
fig.set_size_inches(16, 6)

ax.stem(x, y, label="Temp.")


# set tick and tick labels
ax.set_xticks(x[::1])
ax.set_xticklabels(x[::1], rotation=90)

# set limit on x-axis
ax.set_xlim(0, 10)

# set title
ax.set_title('Temperature')


plt.grid(False)
plt.show()
```
