### **ErrorBar**
Physics Exp. 09 Error Analysis

```python
x = [40, 20, 10, 5, 2.5]

y = [50.5, 24.4, 11.7, 5.55, 2.9]

# find slop from x and y
m = np.polyfit(x, y, 1)[0]
# polyfit function used for find slop

#  find c
c = np.polyfit(x, y, 1)[1]

# find y_the
# y_the = m * x + c
y_the = np.polyval([m, c], x)

# find error
error = np.std(y) / np.sqrt(len(x))

# plot
fig, ax = plt.subplots(2, 1)

fig.set_size_inches(16, 8.27)

ax[0].plot(x, y, 'o-', label='Experimental Data')
ax[0].plot(x, y_the, 'r-', label='Pridicted Data')
ax[0].legend()
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_ylim(0, 10)
ax[0].set_xlim(2, 10)

ax[1].errorbar(x, y, yerr=error, fmt='o-', label='Experimental Data with error', linewidth=2, capsize=6)
ax[1].legend()
ax[1].set_xlabel('x')
ax[1].set_ylabel('y' + 'Error')
ax[1].set_ylim(0, 60)
ax[1].set_xlim(0, 60)


plt.show()
```

## **Physics Exp.13**

```python
# Given data for the example
frequency =600  # Hz
peak_voltage = 8  # Volts

period = 1 / frequency  # seconds

time = np.linspace(0, 5 * period, 1555)  # two cycles for better visualization

# Calculate the waveform
voltage = peak_voltage * np.sin(2 * np.pi * frequency * time)

# Plot the waveform
fig, ax = plt.subplots(figsize=(10, 6))
# fig.set_size_inches(16, 8.27)
ax.plot(time, voltage, label='Voltage (V)')

ax.set_title('Voltage Waveform')
ax.set_xlabel('Time (s)')
ax.set_ylabel('Voltage (V)')

ax.axhline(y=0, color='k', linestyle='--', linewidth=0.8)

ax.axhline(y=peak_voltage, color='r', linestyle='--', linewidth=0.8, label='Peak Voltage (5V)')

ax.axhline(y=-peak_voltage, color='r', linestyle='--', linewidth=0.8)
ax.axhline(y=peak_voltage/np.sqrt(2), color='g', linestyle='--', linewidth=0.8, label='RMS Voltage (~3.54V)')
ax.axhline(y=-peak_voltage/np.sqrt(2), color='g', linestyle='--', linewidth=0.8)

ax.legend()
ax.set_xticks([])
ax.set_yticks([])

# show plot
plt.show()
```

## plot_surface(X, Y, Z)

```python
# cm is a module in Matplotlib that provides a variety of colormaps for plotting.
from matplotlib import cm


# Make data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

'''
X = np.arange(-5, 5, 0.25) creates an array of values from -5 to 5 with a step size of 0.25 for the x-coordinates.
Y = np.arange(-5, 5, 0.25) creates an array of values from -5 to 5 with a step size of 0.25 for the y-coordinates.
X, Y = np.meshgrid(X, Y) generates a mesh grid from the X and Y arrays, creating coordinate matrices.
R = np.sqrt(X**2 + Y**2) calculates the radius R at each point on the grid.
Z = np.sin(R) calculates the sine of the radius R at each point on the grid, producing the z-coordinates for the surface.
'''

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap=cm.Reds)
'''
The vmin=Z.min() * 2 sets the minimum value for color scaling,
and cmap=cm.Blues applies the 'Blues' colormap to the surface.
'''
ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

## plot_trisurf(x, y, z)

```python
'''
n_radii = 8 sets the number of radii.
n_angles = 36 sets the number of angles.
'''
n_radii = 8
n_angles = 36

# Make radii and angles spaces
'''
radii = np.linspace(0.125, 1.0, n_radii) generates 8 radii values linearly spaced between 0.125 and 1.0.
angles = np.linspace(0, 2*np.pi, n_angles, endpoint=False)[..., np.newaxis] generates 36 angle values linearly spaced between 0 and
2𝜋(excluding the endpoint) and reshapes them into a column vector.
'''
radii = np.linspace(0.125, 1.0, n_radii)
angles = np.linspace(0, 2*np.pi, n_angles, endpoint=False)[..., np.newaxis]

# Convert polar (radii, angles) coords to cartesian (x, y) coords.
# converts the polar coordinates to Cartesian x-coordinates and adds a central point at 0.
x = np.append(0, (radii*np.cos(angles)).flatten())
y = np.append(0, (radii*np.sin(angles)).flatten())
z = np.sin(-x*y)

# Plot
fig, ax = plt.subplots(subplot_kw={'projection': '3d'})
ax.plot_trisurf(x, y, z, vmin=z.min() * 2, cmap=cm.Blues)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

## voxels([x, y, z], filled)

```python
# Prepare some coordinates
x, y, z = np.indices((8, 8, 8))

# Draw cuboids in the top left and bottom right corners
cube1 = (x < 3) & (y < 3) & (z < 3)
cube2 = (x >= 5) & (y >= 5) & (z >= 5)

# Combine the objects into a single boolean array
voxelarray = cube1 | cube2

# Plot
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.voxels(voxelarray, edgecolor='k')

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```

# **Irregular Gridded Data Plot**

### tricontour(x, y, z)

```python
# make data:
np.random.seed(1)
x = np.random.uniform(-3, 3, 256)
y = np.random.uniform(-3, 3, 256)
z = (1 - x/2 + x**5 + y**3) * np.exp(-x**2 - y**2)
levels = np.linspace(z.min(), z.max(), 7)

# plot:
fig, ax = plt.subplots()

ax.plot(x, y, 'o', markersize=2, color='lightgrey')
ax.tricontour(x, y, z, levels=levels)

ax.set(xlim=(-3, 3), ylim=(-3, 3))

plt.show()
```
